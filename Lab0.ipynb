```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/AIT TUS/Adv ML AI 2023/Lab0/combined_cycle_power_plant.csv")

df.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

y = df[['PE']]
x = df.drop(['AP', 'RH', 'PE'], axis='columns')

# split into training and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
print(x_train.shape)
print(x_test.shape)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

print('intercept', model.intercept_)
print('coefficients', model.coef_)
print('R squared', model.score(x_train, y_train))
```

```python
y_hat = pd.DataFrame(
  {'PE': model.predict(x_test).flatten()}, index=y_test.index)

# convert the prediction outputs to DataFrame
print(mean_squared_error(y_test, y_hat, squared=False))
```

```python
import plotly.express as px
```

```python
y_hat["Flag"]='Predict'
y_test["Flag"]='Real'
v_hat = pd.concat((x_test, y_hat), axis=1)
v_real = pd.concat((x_test, y_test), axis=1)
df_view = pd.concat((v_hat, v_real), axis=0)

px.scatter_3d(df_view, x='AT', y='V', z='PE', color='Flag')
```
